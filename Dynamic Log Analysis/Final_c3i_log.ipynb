```python
def ext_transition(C,C_next,C_at_prev,C_at_next):
    transitions=[]
    predecessasor_candidates=[]
    predecessasor_overlap=[]
    for i in range(len(C)):
        c1=C[i]
        c1_at_prev=C_at_prev[i]
        successor_candidates=[]
        successor_overlap=0.0
        for j in range(len(C_next)):
            c11=C_next[j]
            c11_at_next=C_at_next[j]
            overlap1 = overlap(c1,c11, c1_at_prev,c11_at_next)
            if overlap1 > 0.2:
                successor_candidates.append(c11)
                successor_overlap+=overlap1
                predecessasor_candidates.append(c1)
                predecessasor_overlap.append(overlap1)
        if successor_overlap > 0.7:
            transitions.append([c1,"+",successor_candidates])
    for j in range(len(C_next)):
        if predecessasor_overlap[j] > 0.7:
            transitions.append([predecessasor_candidates[j],"----------------->>>>>>",C_next[j]])
    return transitions
transitions=ext_transition(C,C_next,C_at_prev,C_at_next)
print(len(transitions))
```

```python
def overlap(C,C_next,C_at_prev,C_at_next):
    s=(C.intersection(C_at_next)).union(C_next.intersection(C_at_prev))
    s1=C.union(C_next)
    overlap=float(len(s))/float(len(s1))
    return overlap
```
