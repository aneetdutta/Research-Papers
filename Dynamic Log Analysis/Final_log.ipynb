```python
import numpy as np
import Levenshtein
import datetime

def listToStringWithoutBrackets(list1):
    return str(list1).replace('[','').replace(']','')
def listToStringWithoutComma(list1):
    return str(list1).replace('\'','').replace('\'','')


#df = pd.DataFrame(columns=['Line'])
c=0
time_window=100000
log1=[]
with open("ApacheAccess.log" ,'r') as f:
    for line in f:
        if "GET" in line:
            word=line.split("\"")
            st=word[1].replace("HTTP/1.1","")
            #df.loc[c]=st
            if "GET /sites/default/files/styles/thumbnail/public/" in st:
                st="GET /sites/default/files/styles/thumbnail/public/"
                log1.append(st)
               # print(st)
                continue
            if "GET /sites/default/files/styles/medium/public/" in st:
                st="GET /sites/default/files/styles/medium/public/"
                log1.append(st)
                #print(st)
                continue
                
            if "GET /sites/default/files/" in st:
                word1=st.partition("_")
                #print(word1[0])
                log1.append(word1[0])
                continue
            
            log1.append(st)
            #print(st)
            c=c+1
            if(c==time_window):
                break
           # print(st)
        if "POST" in line:
            word=line.split("\"")
            st=word[1].replace("HTTP/1.1","")
            #df.loc[c]=st
            log1.append(st)
           # print(st)
            c=c+1
            if(c==time_window):
                break
       
           # print(word[1])
#cluster_1(log1)
#for i in range(len(log1)):
 #   print(log1[i])
def clustering(log1):
    cand=[]
    cand.append(log1[0])
    #print(cand)
    C=[]
    C.append({log1[0]})
    interval=200

    #print(j)
       
    for i in range(1,interval):
    #print(cand)
        d=[]
        for j in range(len(cand)): 
            d.append(Levenshtein.distance(cand[j],log1[i]))
        #print(d)
            minimum=min(d)
            maximum=max(d)
        #print(d)
        if minimum>20:
        #print(len(cand))
            cand.append(log1[i])
            C.append({log1[i]})
            
        else:
            ind=d.index(minimum)
            C[ind].add(log1[i])
    return C,cand

C,cand=clustering(log1[0:200])
C_next,cand_next=clustering(log1[200:400])
C_next_next,cand_next_next=clustering(log1[400:600])
current_time=datetime.datetime.now()
#print(cand)

for i in range(len(C_next)):
    file_out=str("cluster")+str(i+1)+".csv"
    A=[]
    A.append(str(current_time))
    A.append(len(C_next[i]))
    A1=listToStringWithoutBrackets(A)
    A2=listToStringWithoutComma(A1)
    thefile = open(file_out, 'w')
    print>>thefile, A2
    print("----------*********-----------******----------*******----------*********-----------******----------*******----------")
    print(C_next[i])
print(len(C))  
print(len(C_next))
print(len(C_next_next))
```

```python
def overlap(C,C_next,C_at_prev,C_at_next):
    s=(C.intersection(C_at_next)).union(C_next.intersection(C_at_prev))
    s1=C.union(C_next)
    overlap=float(len(s))/float(len(s1))
    return overlap
```

```python
def ext_transition(C,C_next,C_at_prev,C_at_next):
    transitions=[]
    predecessasor_candidates=[]
    predecessasor_indexes=[]
    predecessasor_overlap=[]
    trans_len = []
    for i in range(len(C)):
        c1=C[i]
        c1_at_prev=C_at_prev[i]
        
        successor_candidates=[]
        successor_indexes = []
        successor_overlap=0.0
        
        for j in range(len(C_next)):
         
            c11=C_next[j]
            c11_at_next=C_at_next[j]
            overlap1 = overlap(c1,c11, c1_at_prev,c11_at_next)
            
            if overlap1 > 0.2:
                successor_candidates.append(c11) #c11 is the successor_cand of cluste i(fromer time step)
                successor_indexes.append(j) 
                successor_overlap+=overlap1
                
                predecessasor_candidates.append(c1) #c1 is the pred_cand of the cluster j(later time step)
                predecessasor_overlap.append(overlap1)
                predecessasor_indexes.append(j)
        
        if successor_overlap > 0.7:
            #print(i,"---->>>",len(successor_candidates))
            #print(len(successor_candidates[0]))
            #transitions.append([c1,"+",successor_candidates])
            trans_len.append([len(c1),"-->",len(successor_candidates)])
            transitions.append([i,"-->",successor_indexes])
            #time_list.append([len(c1),"-->",len(successor_candidates[0])])
    for i in range(len(predecessasor_overlap)):
        if predecessasor_overlap[i] > 0.7:
            transitions.append([predecessasor_indexes[i],"-->",i])
            trans_len.append([len(predecessasor_candidates),"-->",len(C_next[i])])
            #time_list.append([len(predecessasor_candidates[i]),"-->",len(C_next[i])])
    #tm_lst.append([time_list])
    return transitions,trans_len
transitions,trans_len=ext_transition(C,C_next,C_at_prev,C_at_next)
print(transitions)
print(trans_len)
```
